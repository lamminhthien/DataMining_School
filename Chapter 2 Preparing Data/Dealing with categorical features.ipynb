##  Add pandas to create dataframe from csv

```python
import pandas as pd
```

## Read CSV Files

```python
sales = pd.read_csv(
    'sales_data.csv',
    parse_dates=['Date'])

sales.head()
```

## Encoding categorical features
<br/>
One-hot encoding

```python
pd.get_dummies(sales,columns=['Country'],prefix='C')
```

# Dummy encoding

```python
pd.get_dummies(sales,columns=['Country'],drop_first=True,prefix='C')
```

# One-hot vs dummies

* One-hot encoding: Explainable features
* Dummy encoding: Necessary information without duplication


# Limiting your coulmns

```python
counts = sales['Country'].value_counts()
print(counts)
```

* Find index of sample that have number of country less than 5
* Convert theme to Other

```python
mask = sales['Country'].isin(counts[counts<5].index)
sales['Country'][mask] = 'Other'
```
